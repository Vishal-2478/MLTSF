```python
"""Implementation of Moving Average"""
```

```python
#Function to Calculate Moving Average

def calculate_moving_average(data, k):
    moving_averages = []
    for i in range(len(data) - k + 1):
        window = data[i : i + k]
        window_average = sum(window) / k
        moving_averages.append(window_average)
    return moving_averages
```

```python
#Sales data for few days
sales = [120, 130, 125, 140, 150, 145, 160, 155, 150, 170,
         165, 160, 175, 180, 190, 200, 195, 185, 180, 175,]

"""In this example we are taking window size as 7
This means we take 7 consecutive days at a time to compute the average."""
window_size = 7 

# 1st Order Moving Average
first_order = calculate_moving_average(sales, window_size)

# 2nd Order Moving Average (apply moving average on first order result)
second_order = calculate_moving_average(first_order, window_size)

print("Original Sales:", sales, "\n")

print("------ 1st Order Moving Average ------")
for i, ma in enumerate(first_order, start=window_size):
    print(f"Day {i}: 7-Day 1st Order MA = {ma:.2f}")

print("\n------ 2nd Order Moving Average ------")
for i, ma in enumerate(second_order, start=2*window_size - 1):
    print(f"Day {i}: 7-Day 2nd Order MA = {ma:.2f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5, 4))

# Plot original sales
plt.plot(range(1, 21), sales, marker='o', label="Original Sales")

# Plot 1st order MA (starts at day 7)
plt.plot(range(7, 21), first_order, marker='s', label="1st Order MA (7-day)")

# Plot 2nd order MA (starts at day 13)
plt.plot(range(13, 21), second_order, marker='^', label="2nd Order MA (7-day)")

# Graph details
plt.xlabel("Day")
plt.ylabel("Sales")
plt.title("Sales Data with 1st & 2nd Order Moving Averages")
plt.legend()
plt.grid(True)     
plt.show()
```
